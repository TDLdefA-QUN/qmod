```python
import sys
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy.io import loadmat

import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.animation import FuncAnimation
%matplotlib inline

from sklearn.model_selection import train_test_split
from IPython.display import HTML

# Add the architecture path for the DenseEncoderDecoder and NMSE
sys.path.append("../tools/")
from tools.cubes2npy import CubeFiles
```

```python
cubes=CubeFiles()
cubes.read_cube(fname="../data/density.0000000001.cube")
cubes.read_cube(fname="../data/density.0000000010.cube")
cube_data=cubes.density_as_numpy_3d()
plt.imshow(cube_data[:,:,50,0])
plt.show()
fig = plt.figure()
ax = fig.add_subplot(111)
im = ax.imshow(cube_data[:,:,50,0], cmap='jet')

# update the frames based on the parameter i
def animate(i):
    im.set_array(cube_data[:,:,50,i])
    return [im]

# run the animation
animation = FuncAnimation(fig, animate, frames=cubes.no_cubes_read, interval=1)
plt.close()
HTML(animation.to_jshtml())
```
